```python
training_file_id="file-RE9xKLUxUHANsYFxur9nS3"
```

```python
suffix_name="GPT finetune model"
```

```python
client.fine_tuning.jobs.create(
    training_file=training_file_id,
    model="gpt-4.1-2025-04-14",
    suffix=suffix_name

)
```

```python
client.fine_tuning.jobs.list()
```

```python
for job in client.fine_tuning.jobs.list():
  print(job.fine_tuned_model)
```

```python
for job in client.fine_tuning.jobs.list():
  print(job.fine_tuned_model)
```

Job status

```python
import time

job_id = 'ftjob-TBIgDyDmOnk7eVrqVFCcgNrX'  # Replace with your actual job ID

while True:
  job = client.fine_tuning.jobs.retrieve(job_id)
  print(f"Job status: {job.status}")

  if job.status in ['succeeded', 'failed', 'cancelled']:
    break

  time.sleep(30)  # Wait for 5 seconds before checking again

if job.status == 'succeeded':
  print(f"Fine-tuning job succeeded! Fine-tuned model: {job.fine_tuned_model}")
elif job.status == 'failed':
  print(f"Fine-tuning job failed. Error: {job.error}")
elif job.status == 'cancelled':
  print(f"Fine-tuning job cancelled.")
```

```python
from config_langgraph import CONTENT_LENGTH, LANGUAGE

# Read CSV without header


df = pd.read_csv(
    "./03_article_writer_expand.csv", header=None, delimiter="|", encoding="utf8"
)



print(f"Number of rows: {len(df)}")



# Process each row


count = 0


errors = []


for index, row in df.iterrows():

    row_number = index + 1


    for col_name, value in row.items():

        if col_name == 7:


            # Attempt to parse the content as JSON

            content_data = ast.literal_eval(value)

            if isinstance(content_data, dict):

                content = content_data.get("content", "No content found")

            else:

                content = "Invalid content format"


            language = detect(content)

            if language == "fr":
                language = LANGUAGE
                count += 1
            article_length = len(content.split())

            if article_length < CONTENT_LENGTH:
                errors.append(
                    f"Row {row_number} in column {col_name} has insufficient content length: {article_length} words"
                )
            else:

                print(f" {row_number}: {article_length} words OK in {language} OK")
```
